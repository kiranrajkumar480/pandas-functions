```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\\Users\\KIRAN\\Downloads\\8-Queens-Problem-Genetic-algorithm-Kirankannan\\Desktop\\directors\\director.csv")

# Bar chart to show the number of nominations or winners by gender
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Number of Nominations or Winners by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Bar chart to show the number of nominations or winners by race
race_counts = df['Race'].value_counts()
plt.bar(race_counts.index, race_counts.values)
plt.title('Number of Nominations or Winners by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()

# Bar chart to show the number of nominations or winners by director
director_counts = df['Director(s)'].value_counts().head(10)
plt.bar(director_counts.index, director_counts.values)
plt.title('Number of Nominations or Winners by Director (Top 10)')
plt.xlabel('Director')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

# Line chart to show the trend in nominations or winners over time
year_counts = df['Year'].str.split('/', expand=True)[0].value_counts().sort_index()
plt.plot(year_counts.index, year_counts.values)
plt.title('Trend in Nominations or Winners over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='center')
plt.show()

# Stacked bar chart to show the breakdown of nominations or winners by gender and race
gender_race_counts = df.groupby(['Gender', 'Race']).size().unstack()
gender_race_counts.plot(kind='bar', stacked=True)
plt.title('Breakdown of Nominations or Winners by Gender and Race')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Horizontal bar chart to show the number of nominations or winners by film
film_counts = df['Film'].value_counts().head(10)
plt.barh(film_counts.index[::-1], film_counts.values[::-1])
plt.title('Number of Nominations or Winners by Film (Top 10)')
plt.xlabel('Count')
plt.ylabel('Film')
plt.show()

# Pie chart to show the proportion of nominations or winners by gender
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Nominations or Winners by Gender')
plt.show()

# Pie chart to show the proportion of nominations or winners by race
race_counts = df['Race'].value_counts()
plt.pie(race_counts.values, labels=race_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Nominations or Winners by Race')
plt.show()

# Heatmap to show the number of nominations or winners by year and gender
year_gender_counts = df.groupby(['Year', 'Gender']).size().unstack().fillna(0)
sns.heatmap(year_gender_counts, cmap='Blues')
plt.title('Number of Nominations or Winners by Year and Gender')
plt.xlabel('Gender')
plt.ylabel('Year')
plt.show()
```
