```python
top_10_runs = df.sort_values('Runs', ascending=False).head(10)

sns.countplot(x='Runs', data=top_10_runs)
plt.show()

# Filter
top_10_matches = df.sort_values('Mat', ascending=False).head(10)

sns.countplot(x='Mat', data=top_10_matches)
plt.show()

# Filter based on SR
top_10_sr = df.sort_values('SR', ascending=False).head(10)
sns.countplot(x='SR', data=top_10_sr)
plt.show()
```

```python
sns.barplot(x='Player', y='Runs', data=top_10_runs)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='Mat', y='Runs', data=top_10_runs)
plt.show()
```

```python
sns.boxplot(x='Player', y='SR', data=top_10_sr)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.violinplot(x='Player', y='Runs', data=top_10_runs)
plt.xticks(rotation=90)
plt.show()
```

```python
# strike rates of the top 10 players
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Player', y='SR', data=top_10_sr)
plt.title('Top 10 Players by Strike Rate')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# matches played by the top 10 players
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Mat', data=top_10_matches)
plt.title('Top 10 Players by Matches Played')
plt.xlabel('Player')
plt.ylabel('Matches Played')
plt.xticks(rotation=45)
plt.show()
```

```python
# relationship between runs and strike rate for the top 10 players
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Runs', y='SR', data=top_10_runs)
plt.title('Relationship Between Runs and Strike Rate')
plt.xlabel('Runs')
plt.ylabel('Strike Rate')
plt.show()
```

```python
# distribution of strike rates for the top 10 players
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.violinplot(x='Player', y='SR', data=top_10_sr)
plt.title('Distribution of Strike Rates for Top 10 Players')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# correlation matrix for the top 10 players
corr_matrix = top_10_runs[['Runs', 'BF', 'SR', '100', '50', '4s', '6s']].corr()
sns.set(style='white')
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Top 10 Players')
plt.show()
```
